```python
!pip install youtube-transcript-api

from youtube_transcript_api import YouTubeTranscriptApi
from youtube_transcript_api.formatters import TextFormatter
from IPython.display import display, HTML

def create_copy_button_with_text(text):
    display(HTML(f"""
    <textarea id="text" style="display:none;">{text}</textarea>
    <button onclick="copyText()">Copy Transcript</button>
    <script>
        function copyText() {{
            var copyText = document.getElementById('text');
            copyText.style.display = 'block';
            copyText.select();
            document.execCommand('copy');
            copyText.style.display = 'none';
            alert("Transcript copied to clipboard!");
        }}
    </script>
    """))
    # Menampilkan transkrip
    print("Transkrip Video:\n")
    print(text)

# Specify the video ID
video_id = 'ErnWZxJovaM'  # Ganti dengan video ID yang diinginkan

# Retrieve the transcript using the Indonesian auto-generated transcript
transcript = YouTubeTranscriptApi.get_transcript(video_id, languages=['id','en'])

# Format the transcript as text
formatter = TextFormatter()
text = formatter.format_transcript(transcript)

# Display the copy button with the transcript
create_copy_button_with_text(text)
```
